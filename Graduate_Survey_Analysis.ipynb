```python
# Clean data (handle missing values, filter employed respondents)
df_employed = df[df['Current Status'].str.contains('Employed', na=False)]
df_employed['Annual Salary (pre-tax)'] = df_employed['Annual Salary (pre-tax)'].str.replace('$', '').str.replace(',', '').astype(float)  # Convert salary to numeric

# -----------------------------------------
# 1. Key Metrics Calculation
# -----------------------------------------
total_respondents = len(df)
employment_rate = len(df_employed) / total_respondents * 100
further_education_rate = (df['Current Status'].str.contains('Pursuing further education').sum() / total_respondents) * 100
top_industry = df_employed['Industry'].mode()[0]
avg_salary = df_employed['Annual Salary (pre-tax)'].median()  # Use median to reduce skew
job_satisfaction_avg = df_employed['Job Satisfaction'].mean()

# -----------------------------------------
# 2. Visualizations
# -----------------------------------------
plt.style.use('seaborn')

# Figure 1: Employment Status Distribution
fig1, ax1 = plt.subplots(figsize=(8, 5))
status_counts = df['Current Status'].value_counts()
status_counts.plot(kind='bar', ax=ax1, color='skyblue')
ax1.set_title('Employment Status Distribution', fontsize=14)
ax1.set_xlabel('Status')
ax1.set_ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()

# Figure 2: Top Industries
fig2, ax2 = plt.subplots(figsize=(8, 5))
industry_counts = df_employed['Industry'].value_counts().head(5)
industry_counts.plot(kind='pie', autopct='%1.1f%%', ax=ax2, colors=sns.color_palette('pastel'))
ax2.set_title('Top 5 Industries for Graduates', fontsize=14)
ax2.set_ylabel('')

# Figure 3: Salary Distribution by Field
fig3, ax3 = plt.subplots(figsize=(10, 6))
sns.boxplot(data=df_employed, x='Field of Study', y='Annual Salary (pre-tax)', palette='Set2')
ax3.set_title('Salary Distribution by Field of Study', fontsize=14)
ax3.set_xlabel('Field of Study')
ax3.set_ylabel('Annual Salary (USD)')
plt.xticks(rotation=45)

# Figure 4: Skills Gaps
skills_gap = df['Which skills do you wish you had developed more?'].value_counts()
fig4, ax4 = plt.subplots(figsize=(8, 4))
skills_gap.plot(kind='barh', ax=ax4, color='lightgreen')
ax4.set_title('Most Requested Skills for Improvement', fontsize=14)
ax4.set_xlabel('Count')

# -----------------------------------------
# 3. Generate Report Text
# -----------------------------------------
report_text = f"""
**Executive Summary Report**

1. **Employment Rate**: {employment_rate:.1f}% of graduates are employed.
2. **Further Education**: {further_education_rate:.1f}% are pursuing additional degrees.
3. **Top Industry**: {top_industry} employs the most graduates.
4. **Median Salary**: ${avg_salary:,.0f} (varies by field; see Figure 3).
5. **Job Satisfaction**: {job_satisfaction_avg:.1f}/5 average rating.

**Recommendations**:
- Address skills gaps in {skills_gap.index[0]} (Figure 4).
- Strengthen partnerships with {top_industry} employers.
"""

print(report_text)
plt.show()
```
